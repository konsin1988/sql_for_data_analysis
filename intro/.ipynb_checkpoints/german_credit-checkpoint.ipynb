```python
import pandas as pd
import sqlite3

df = pd.read_csv('../german_credit_augmented.csv').assign(contract_dt = lambda x: pd.to_datetime(x['contract_dt']))
```

```python
df.head()
```

```python
df.info()
```

```python
connection = sqlite3.connect('../german_credit.db')
```

```python
df.to_sql('german_credit', connection, index=False, if_exists='replace')
```

```python
query = '''
    SELECT * FROM german_credit t
'''
```

```python
pd.read_sql_query(query, connection)
```

```python
# sql first day of month
query = '''
    SELECT contract_dt,
        date(contract_dt, 'start of month') AS month_start
    FROM german_credit
    ORDER BY contract_dt
'''
pd.read_sql_query(query, connection, dtype={'month_start': 'datetime64[ns]'})
```
